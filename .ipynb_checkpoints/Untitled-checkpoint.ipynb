# Profilowanie obciążenia korytarzy transportowych A1/S8 w celu optymalizacji pracy centrum dystrybucyjnego

**Problem badawczy:**
Jak cykl tygodniowy oraz zmienność kalendarzowa wpływają na ruch ciężarowy na korytarzach A1 i S8? Celem analizy jest identyfikacja wzorców ruchu na Węźle Łódź Południe, stanowiącym główne okno logistyczne dla centrum w Rzgowie, oraz ocena ryzyka zatorów.

**Założenie analityczne (Scope Definition):**
W celu zapewnienia wysokiej jakości wniosków operacyjnych, zbiór danych z systemu e-TOLL został zawężony wyłącznie do dni roboczych (poniedziałek-piątek). Pozwala to na wyeliminowanie szumu informacyjnego z weekendów i skupienie się na potokach ciężarowych istotnych dla procesów magazynowych.

# Analiza Dzienna

### Import bibliotek i konfiguracja środowiska

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Paleta IBCS do raportowania biznesowego
ibcs_palette = ['#000000', '#333333', '#555555', '#777777', '#999999', '#AAAAAA', '#BBBBBB', '#CCCCCC', '#DDDDDD', '#EEEEEE']
sns.set_style("ticks")

# Wczytanie danych
df_drogi_dzienny = pd.read_csv('data/pojazdy_wg_numeru_drogi_dzienna.csv', sep=',')

# Konwersja dat
df_drogi_dzienny['Data'] = pd.to_datetime(df_drogi_dzienny['Data'])

# Mapowanie dni tygodnia
dni_tygodnia_mapa = {
    'Monday': 'Poniedziałek', 'Tuesday': 'Wtorek', 'Wednesday': 'Środa',
    'Thursday': 'Czwartek', 'Friday': 'Piątek', 'Saturday': 'Sobota', 'Sunday': 'Niedziela'
}
df_drogi_dzienny['Dzień_tygodnia_pl'] = df_drogi_dzienny['Data'].dt.day_name().map(dni_tygodnia_mapa)

kolejnosc_dni = ['Poniedziałek', 'Wtorek', 'Środa', 'Czwartek', 'Piątek', 'Sobota', 'Niedziela']
df_drogi_dzienny['Dzień_tygodnia_pl'] = pd.Categorical(
    df_drogi_dzienny['Dzień_tygodnia_pl'], 
    categories=kolejnosc_dni, 
    ordered=True
)
```

### Definicja Zakresu Analizy

```python
# Filtrowanie do dni roboczych
df_robocze = df_drogi_dzienny[df_drogi_dzienny['Dzień_tygodnia_pl'].isin(['Poniedziałek', 'Wtorek', 'Środa', 'Czwartek', 'Piątek'])].copy()

# Usunięcie pustych kategorii (Sobota, Niedziela), aby nie psuły osi wykresów
df_robocze['Dzień_tygodnia_pl'] = df_robocze['Dzień_tygodnia_pl'].cat.remove_unused_categories()

print("Struktura danych operacyjnych (dni robocze):")
display(df_robocze.head())
```

### Analiza Zależności tras (Korelacja)

```python
# Korelacja Dzień vs Ruch (spadek/wzrost w trakcie tygodnia roboczego)
df_robocze['Dzien_Num'] = df_robocze['Dzień_tygodnia_pl'].cat.codes
korelacja_dzien = df_robocze['Dzien_Num'].corr(df_robocze['Wartość'])
print(f"Korelacja upływu tygodnia roboczego do natężenia ruchu: {korelacja_dzien:.2f}")

# Heatmapa A1 vs S8
plt.figure(figsize=(6, 4))
df_pivot = df_robocze.pivot(index='Data', columns='Numer drogi', values='Wartość')
sns.heatmap(df_pivot.corr(), annot=True, cmap='Greys', vmin=0.8, vmax=1.0)
plt.title('Współczynnik korelacji: Korytarz A1 vs S8', pad=15, fontweight='bold')
plt.show()
```

**Wniosek biznesowy:** Brak silnej korelacji (-0.13) oraz wyrównane poziomy na wykresie słupkowym dowodzą, że ruch roboczy jest wysoce stabilny. Rekomenduje to wdrożenie tzw. płaskiego harmonogramu pracy na rampach w Rzgowie, bez konieczności codziennego, kosztownego manipulowania liczbą personelu.  
**Wniosek biznesowy:** Ekstremalnie wysoka korelacja (0.98) udowadnia, że trasy A1 i S8 reagują identycznie na sytuację makroekonomiczną i drogową. Korytarze te nie dają możliwości dywersyfikacji ryzyka – w przypadku zatorów firma musi przygotować się na opóźnienia z obu kierunków jednocześnie.

### Ocena Przepustowości Dziennej (Barplot)

```python
# Agregacja do średnich
analiza_trasy = df_robocze.groupby(['Dzień_tygodnia_pl', 'Numer drogi'], observed=True)['Wartość'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=analiza_trasy, x='Dzień_tygodnia_pl', y='Wartość', hue='Numer drogi', 
            palette=[ibcs_palette[1], ibcs_palette[4]])
plt.title('Średnie natężenie ruchu roboczego: A1 vs S8', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Średnia dzienna liczba pojazdów ciężkich')
plt.xlabel('Dzień roboczy')
sns.despine()
plt.legend(title='Trasa', bbox_to_anchor=(1.02, 1), loc='upper left', frameon=False)
plt.show()
```

### Zarządzanie Ryzykiem i Zmienność (Boxplot)

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_robocze, x='Numer drogi', y='Wartość', hue='Numer drogi', 
            palette=[ibcs_palette[1], ibcs_palette[4]], legend=False, width=0.5)

plt.title('Rozkład ruchu w dni robocze: Identyfikacja anomalii', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Dzienna liczba pojazdów ciężkich')
plt.xlabel('Numer drogi')
sns.despine()
plt.show()
```

**Wniosek biznesowy:** Standardowy ruch jest wysoce przewidywalny (zwarte "pudełko"). Zaobserwowane liczne wartości odstające to anomalie wynikające głównie z dni ustawowo wolnych (święta państwowe) wypadających w środku tygodnia roboczego. To otwiera drogę do stworzenia systemu wczesnego ostrzegania o nagłych spadkach przepustowości
